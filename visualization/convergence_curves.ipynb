```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
```

## BaseLine

Batch size 16

```python
baseline_bs16 = '../results/baseline_bs16_patience10_epochs50/'
res1 = pd.read_csv(baseline_bs16 + 'metrics_run1.csv', index_col=0).replace(0, np.NaN).dropna()
res2 = pd.read_csv(baseline_bs16 + 'metrics_run2.csv', index_col=0).replace(0, np.NaN).dropna()
res3 = pd.read_csv(baseline_bs16 + 'metrics_run3.csv', index_col=0).replace(0, np.NaN).dropna()
res4 = pd.read_csv(baseline_bs16 + 'metrics_run4.csv', index_col=0).replace(0, np.NaN).dropna()
res5 = pd.read_csv(baseline_bs16 + 'metrics_run5.csv', index_col=0).replace(0, np.NaN).dropna()
```

```python
plt.plot(res1['train_loss'], label="Run 1")
plt.plot(res2['train_loss'], label="Run 2")
plt.plot(res3['train_loss'], label="Run 3")
plt.plot(res4['train_loss'], label="Run 4")
plt.plot(res5['train_loss'], label="Run 5")
plt.title("Evolution of Training Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.savefig("base_bs16_loss")
```

```python
plt.plot(100 * res1['val_accuracy'], label="Run 1")
plt.plot(100 * res2['val_accuracy'], label="Run 2")
plt.plot(100 * res3['val_accuracy'], label="Run 3")
plt.plot(100 * res4['val_accuracy'], label="Run 4")
plt.plot(100 * res5['val_accuracy'], label="Run 5")
plt.title("Evolution of Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy [%]")
plt.legend()
plt.savefig("base_bs16_acc")
```
